$\alpha/p = A\exp(-Bp/E)$

$\alpha d = \log(1 + 1/\gamma)$

$E=V_S/d,\,\gamma=cst.$

$\frac{\log(1+1/\gamma)}{dp} = A\exp\left(-\frac{Bdp}{V_S}\right)$

$-\frac{Bdp}{V_S} = \log\left(\frac{\log(1+1/\gamma)}{Adp}\right)$

$V_S = \frac{Bdp}{\log(dp) +\log(A) - \log\log(1+1/\gamma)}$
